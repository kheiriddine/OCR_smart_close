```python
import fitz  # PyMuPDF
import re
import pdfplumber
from datetime import datetime
from tabula import read_pdf
import pandas as pd
from functools import reduce
import json
import os
from PIL import Image

List_regex101 = [
    # r'...'=raw string
    # \s = un caractère d'espacement (espace, tabulation, etc.)
    # * = quantificateur "0 ou plusieurs fois"
    # [ ] = définit une classe de caractères
    # \d = un chiffre (0-9)
    # , = la virgule littérale
    # + = quantificateur "1 ou plusieurs fois"
    # \. = le point littéral (échappé avec \)
    # \d{2} = exactement 2 chiffres
]

def format_date(date_str):
    """Convertit une date 'DD/MM/YYYY' en objet date puis en format 'YYYY-MM-DD' (ou autre)"""
    try:
        date_obj = datetime.strptime(date_str, '%d/%m/%Y').date()
        return date_obj.strftime('%d/%m/%Y')  #('%d/%m/%Y'le format français) ou bien Format ISO
    except ValueError:
        return date_str

def convert_data_types(data):
    """Convertit les types de données dans le dictionnaire"""
    if 'Date Paiement' in data:
        data['Date Paiement'] = format_date(data['Date Paiement'])

    numeric_fields = ['Num Cycle Paiement', 'Code Banque', 'Code Agence', 'Reference Paiement']
    for field in numeric_fields:
        if field in data:
            try:
                data[field] = int(data[field])
            except ValueError:
                pass

    if 'Total' in data:
        data['Total'] = float(data['Total'].replace(',', '').replace(' ', '').replace('\xa0', ''))

    return data

def extract_payment_info_v0(pdf_path):
    patterns = {
        'Nom Fournisseur': r'Fournisseur:\s*([^\n]+)',
        'Adresse Fournisseur': r'Adresse Fournisseur:\s*([^\n]+)',
        'Cycle Paiement': r'Cycle Paiement:\s*([^\n]+)',
        'Num Cycle Paiement': r'Num Cycle Paiement:\s*(\d+)',
        'Reference Paiement': r'Reference Paiement:\s*(\S+)',
        'Date Paiement': r'Date Paiement:\s*(\d{2}/\d{2}/\d{4})',
        'Methode Paiement': r'Methode Paiement:\s*([^\n]+)',
        'Code Banque': r'Information Banque:\s*(\d+)',
        'Nom Banque': r'Information Banque:\s*\d+\s+([^\n]+?)\s+\d+',
        'Code Agence': r'Information Banque:\s*\d+\s+[^\n]+?\s+(\d+)',
        'Compte Bancaire': r'Compte Bancaire:\s*(\S+)',
        'Total':r'Total\s+[Pp]ayé:\s*([\d.]+)' # r'Total:\s*([\d,]+\.\d{2})'
    }

    doc = fitz.open(pdf_path)
    text = "\n".join(page.get_text("text") for page in doc)
    text = re.sub(r'\s{2,}', ' ', text)

    results = {}
    for field, pattern in patterns.items():
        match = re.search(pattern, text)
        if match:
            results[field] = match.group(1).strip()
    return convert_data_types(results)

def extract_payment_info_v1(pdf_path):
    """Extraction ciblée avec les motifs spécifiés"""
    patterns = {
        'Nom Fournisseur': r'Fournisseur:\s*([^\n]+?)(?=\s+Adresse|\s+Information|$)',
        'Adresse Fournisseur': r'Adresse Fournisseur:\s*([^\n]+?)(?=\s+Cycle|\s+Information|$)',
        'Cycle Paiement': r'Cycle Paiement:\s*([^\n]+)',
        'Num Cycle Paiement': r'Num Cycle Paiement:\s*(\d+)',
        'Reference Paiement': r'Reference Paiement:\s*(\S+)',
        'Date Paiement': r'Date Paiement:\s*(\d{2}/\d{2}/\d{4})',
        'Methode Paiement': r'Methode Paiement:\s*([^\n]+)',
        'Code Banque': r'Information Banque:\s*(\d+)',
        'Nom Banque': r'Information Banque:\s*\d+\s+([^\n]+?)\s+\d+',
        'Code Agence': r'Information Banque:\s*\d+\s+[^\n]+?\s+(\d+)',
        'Compte Bancaire': r'Compte Bancaire:\s*(\S+)',
        'Total':r'Total\s+[Pp]ayé:\s*([\d.]+)' #r'Total:\s*([\d,]+\.\d{2})'
    }
    with pdfplumber.open(pdf_path) as pdf:
        text = "\n".join(page.extract_text() for page in pdf.pages)

    text = re.sub(r'\s{2,}', ' ', text)
    #print(text)

    results = {}
    for field, pattern in patterns.items():
        match = re.search(pattern, text)
        if match:
            results[field] = match.group(1).strip()

    return convert_data_types(results)

def extract_tables(file_path):
    """
    Extrait les tables en supprimant :
    - La dernière ligne
    - Les colonnes 'Unnamed: 0' (index inutiles)
    """
    if file_path.lower().endswith('.pdf'):
        tables = read_pdf(file_path, pages='all', multiple_tables=True)
    else:
        # Conversion image en PDF temporaire
        image = Image.open(file_path)
        pdf_path = "temp_image.pdf"
        image.save(pdf_path, 'PDF')
        tables = read_pdf(pdf_path, pages='all', multiple_tables=True)
        os.remove(pdf_path)  # Suppression du PDF temporaire

    processed_tables = []
    for table in tables:
        if not table.empty:
            processed_table = table.iloc[:-1] if len(table) > 1 else table
            processed_table = processed_table.loc[:, ~processed_table.columns.str.contains('^Unnamed')]
            processed_tables.append(processed_table)
    return processed_tables

def join_tables(tables, how='concat', on=None):
    """
    Joint plusieurs DataFrames selon différentes méthodes
    Args:
        tables (list): Liste de DataFrames pandas
        how (str): Méthode de jointure ('concat', 'merge', 'append')
        on (str): Colonne de jointure (pour les merges)
    Returns:
        DataFrame: Table résultante de la jointure
    """
    if not tables:
        return pd.DataFrame()

    if how == 'concat':
        # Concaténation verticale (empilement)
        return pd.concat(tables, ignore_index=True)

    elif how == 'merge':
        if not on:
            raise ValueError("Must specify 'on' column for merge")
        return reduce(lambda left, right: pd.merge(left, right, on=on), tables)

    elif how == 'append':
        result = tables[0]
        for table in tables[1:]:
            result = result.append(table, ignore_index=True)
        return result
    else:
        raise ValueError(f"Unknown join method: {how}")

def clean_and_convert_table(df):
    """Nettoie et convertit les types de données dans le DataFrame"""
    if 'Reference Paiement' in df.columns:
        df['Reference Paiement'] = df['Reference Paiement'].astype(int)
    if 'Montant Facture' in df.columns:
        df['Montant Facture'] = df['Montant Facture'].str.replace(',', '').astype(float)
    if 'Montant Paye' in df.columns:
        df['Montant Paye'] = df['Montant Paye'].str.replace(',', '').astype(float)
    return df

def main(pdf_folder):
    combined_payment_info = {}
    all_tables = []

    # Parcourir tous les fichiers dans le dossier
    for filename in os.listdir(pdf_folder):
        if filename.lower().endswith('.pdf'):
            pdf_path = os.path.join(pdf_folder, filename)
            payment_info = extract_payment_info_v1(pdf_path)
            tables = extract_tables(pdf_path)

            # Fusionner les informations de paiement
            for key, value in payment_info.items():
                if key in combined_payment_info:
                    if combined_payment_info[key] != value:
                        combined_payment_info[key] = f"{combined_payment_info[key]}, {value}"
                else:
                    combined_payment_info[key] = value

            all_tables.extend(tables)

    if not all_tables:
        print("Aucune donnée n'a pu être extraite")
        return

    # Concaténer tous les DataFrames
    combined_df = join_tables(all_tables, how='concat', on=None)
    combined_df = clean_and_convert_table(combined_df)

    # Enregistrer les informations dans un fichier JSON
    output = {
        'payment_info': combined_payment_info,
        'tables': combined_df.to_dict(orient='records')
    }

    with open('output_pdf_test.json', 'w') as f:
        json.dump(output, f, indent=4)

    print("Informations enregistrées dans output_pdf_test.json")

pdf_folder = "/content/PDFs"
pdf_folder_test="/content/sample_data/SearchablePDFs"
main(pdf_folder)
```

##**Data Extraction from Images**

```python
import re
import os
import pandas as pd
import json
from PIL import Image
import pytesseract
import cv2
import numpy as np
from datetime import datetime

def preprocess_image(image_path):
    """Améliore la qualité de l'image pour une meilleure OCR"""
    img = cv2.imread(image_path)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    thresh = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)[1]
    kernel = np.ones((1, 1), np.uint8)
    cleaned = cv2.morphologyEx(thresh, cv2.MORPH_OPEN, kernel)
    return cleaned

def extract_text_from_image(image_path):
    """Extrait le texte de l'image avec prétraitement"""
    try:
        processed_img = preprocess_image(image_path)
        custom_config = r'--oem 3 --psm 6 -l fra+eng'
        text = pytesseract.image_to_string(processed_img, config=custom_config)
        return text
    except Exception as e:
        print(f"Erreur OCR: {e}")
        return ""

def format_date(date_str):
    """Convertit une date au format 'jj/mm/yyyy' en 'jj/mm/yy'"""
    try:
        date_obj = datetime.strptime(date_str, '%d%m/%Y')
        return date_obj.strftime('%d/%m/%y')
    except ValueError:
        return date_str

def extract_payment_info(text):
    """Extrait les informations principales de paiement"""
    patterns = {
        'Nom Fournisseur': r'Fournisseur:\s*([^\n]+?)(?=\s+Adresse|\s+Information|$)',
        'Adresse Fournisseur': r'Adresse Fournisseur:\s*([^\n]+?)(?=\s+Cycle|\s+Information|$)',
        'Cycle Paiement': r'Cycle Paiement:\s*([A-Z0-9]+)',
        'Num Cycle Paiement': r'Num Cycle Paiement:\s*(\d+)',
        'Reference Paiement': r'Reference Paiement:\s*(\d+)',
        'Date Paiement': r'Date Paiement:\s*(\d{4}/\d{4})',
        'Methode Paiement': r'Methode Paiement:\s*([^\n]+?)(?=\s*Code Banque|$)',
        'Code Banque': r'Information Banque:\s*(\d+)',
        'Nom Banque': r'Information Banque:\s*\d+\s+([a-zA-Z]+)',
        'Code Agence': r'Information Banque:\s*\d+\s+[a-zA-Z]+\s+(\d+)',
        'Compte Bancaire': r'Compte Bancaire:\s*(\*{8}\d{3})',
        'Total': r'Total[:\s]+([\d,.]+)'
    }

    results = {}
    for field, pattern in patterns.items():
        match = re.search(pattern, text)
        if match:
            results[field] = match.group(1).strip()

    # Conversion des types de données
    if 'Nom Banque' in results:
        results['Nom Banque'] = results['Nom Banque'].upper()
    numeric_fields = ['Num Cycle Paiement', 'Reference Paiement', 'Code Banque', 'Code Agence']
    for field in numeric_fields:
        if field in results:
            results[field] = int(results[field])
    if 'Total' in results:
        results['Total'] = float(results['Total'].replace(',', '').replace(' ', '').replace('\xa0', ''))
    if 'Date Paiement' in results:
        results['Date Paiement'] = format_date(results['Date Paiement'])

    return results

def extract_tables_to_dataframe(text):
    """Extrait les informations des tables et les retourne dans un DataFrame"""
    # Pattern pour capturer chaque ligne de table
    table_pattern = r'([A-Za-z0-9]+)\s+(\d{2}/\d{2}/\d{4})\s+(\d+)\s+([\d,]+\.\d{2})\s+([\d,]+\.\d{2})\s+([\d,]+\.\d{2})\s+([\d,]+\.\d{2})'
    matches = re.findall(table_pattern, text)

    # Création du DataFrame
    columns = ['Num facture', 'Date facture', 'Piece', 'Montant Facture', 'Escompte Pris', 'Frais Retard', 'Montant Paye']
    df = pd.DataFrame(matches, columns=columns)

    # Conversion des types de données
    df['Date facture'] = pd.to_datetime(df['Date facture'], format='%d/%m/%Y').dt.strftime('%d/%m/%y')
    df['Piece'] = df['Piece'].astype(int)

    # Nettoyage des valeurs numériques (remplacement des virgules)
    numeric_cols = ['Montant Facture', 'Escompte Pris', 'Frais Retard', 'Montant Paye']
    for col in numeric_cols:
        df[col] = df[col].str.replace(',', '').astype(float)

    df['Num facture'] = df['Num facture'].astype(str)

    return df

def process_image(image_path):
    """Traite l'image et extrait les données structurées"""
    text = extract_text_from_image(image_path)
    if not text:
        return None

    text = re.sub(r'\s+', ' ', text)

    payment_info = extract_payment_info(text)
    df = extract_tables_to_dataframe(text)

    return {
        'payment_info': payment_info,
        'tables': df,
        'total_amount': df['Montant Paye'].sum()
    }

def main():
    image_folder = "/content/Images"
    all_tables = []
    total_amount = 0.0
    combined_payment_info = {}

    # Parcourir tous les fichiers dans le dossier
    for filename in os.listdir(image_folder):
        if filename.lower().endswith(('.png', '.jpg', '.jpeg', '.tiff', '.bmp', '.gif')):
            image_path = os.path.join(image_folder, filename)
            result = process_image(image_path)
            if result:
                all_tables.append(result['tables'])
                total_amount += result['total_amount']

                # Fusionner les informations de paiement
                for key, value in result['payment_info'].items():
                    if key in combined_payment_info:
                        if combined_payment_info[key] != value:
                            combined_payment_info[key] = f"{combined_payment_info[key]}, {value}"
                    else:
                        combined_payment_info[key] = value

    if not all_tables:
        print("Aucune donnée n'a pu être extraite")
        return

    # Concaténer tous les DataFrames
    combined_df = pd.concat(all_tables, ignore_index=True)

    # Vérification du préfixe des numéros de facture
    prefixes = combined_df['Num facture'].str.extract(r'^([A-Za-z]+)')[0].unique()
    same_prefix = len(prefixes) == 1

    # Enregistrer les informations dans un fichier JSON
    output = {
        'payment_info': combined_payment_info,
        'tables': combined_df.to_dict(orient='records'),
        'total_amount': total_amount,
        'same_prefix': same_prefix
    }

    with open('output_images.json', 'w') as f:
        json.dump(output, f, indent=4)

    print("Informations enregistrées dans output_images.json")

if __name__ == "__main__":
    main()
```

```python
import pytesseract
from PIL import Image
import re
import json


def extract_purchase_order_data(text):
    """Extrait les données clés du bon de commande."""
    data = {
        "numero_commande": None,
        "fournisseur": None,
        "date": None,
        "articles": [],
        "montant_total": None
    }

    # 1. Trouver le numéro de commande (ex: "N°CMD-1234")
    cmd_match = re.search(r"(Ref.\s*[:.-]?\s*[A-Z]*-?\d+)", text, re.IGNORECASE)
    if cmd_match:
        data["numero_commande"] = cmd_match.group(1).strip()

    # 2. Trouver le fournisseur (ex: "Fournisseur : XYZ SARL")
    supplier_match = re.search(r"(Information société\s*[:.-]?\s*([A-Z]+))", text, re.IGNORECASE)
    if supplier_match:
        data["fournisseur"] = supplier_match.group(2).strip()

    date_match = re.search(r"(Date\s*[:.-]?\s*(\d{2}[/-]\d{2}[/-]\d{4}))", text)
    if date_match:
        data["date"] = date_match.group(2).strip()

    articles = re.findall(r"(\d+)\s*x?\s*([^\n]+?)\s*(\d+[,.]\d{2})\s*€?", text)
    print(articles)
    for qty, desc, prix_u in articles: #, prix
        data["articles"].append({
            "quantite": int(qty),
            "description": desc.strip(),
            "prix_unitaire": float(prix_u.replace(',', '.')),
            #"prix_TTC":float(prix.replace(',', '.'))
        })

    total_match = re.search(r"(Total\s*[:.-]?\s*(\d+[,.]\d{2})\s*€?)", text, re.IGNORECASE)
    if total_match:
        data["montant_total"] = float(total_match.group(2).replace(',', '.'))

    return data

image_path = "/content/Images/Capture d’écran 2025-04-30 142015.png"
extracted_text = extract_text_from_image(image_path)
print(extracted_text)
purchase_data = extract_purchase_order_data(extracted_text)

with open("bon_commande.json", "w", encoding="utf-8") as f:
    json.dump(purchase_data, f, ensure_ascii=False, indent=4)

print(json.dumps(purchase_data, indent=4, ensure_ascii=False))
```
